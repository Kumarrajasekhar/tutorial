```python
def outliers(df,j):# d is dff and i is column name
    q1=df[j].quantile(0.25)
    q3=df[j].quantile(0.75)
    iqr=q3-q1
    low=q1-(1.5*iqr)
    upp=q3+(1.5*iqr)
    vi=df[(df[j]>=low)&(df[j]<=upp)]
    m=vi[j].mean()
    df[j].apply(lambda x:m if x<=low else m if x>=upp else x)#capping by mean
    
    
    # | is for see outliers and & is for check for remailing valid data.
    
    
```

```python
def outlier(df,i):
    m=round(df[i].mean(),2)# 2nd method of outlier
    sd=round(df[i].std(),2)
    low=m-(3*sd)
    upp=m+(3*sd)
    vi=df[(df[i]>=low)&(df[i]<=upp)]
    m=vi[i].mean()
    df[i].apply(lambda x: m if x<=low else m if x>=upp else x )
```
